```python
#COM EFEITO BLURR

image = cv2.imread('Fig0334.tif')
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 

new_f = cv2.medianBlur(image, 15)

plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(new_f, cv2.COLOR_HSV2RGB)),plt.title('Filtro')
plt.xticks([]), plt.yticks([])
plt.show()

print("Percebe-se que o efeito de filtragem conseguiu eliminar os pequenos elementos dispostos na imagem, porém não sem realizar um efeito de blurr que dificulta a compreensão da imagem e posteriorm análise, sendo necessário etapa posterior para recuperar informações que o filtro aplicado não permite visualizar com perfeição")
```

```python
#COMO DEMONSTRADO EM AULA JUNTO AO MESTRANDO CALANGO

img = cv2.imread('Fig0334.tif')
N = 5 #tamanho da matriz
mascara_esp = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(N,N))
imgfinal= cv2.erode(img,mascara_esp)


figure, saida = plt.subplots(1,2, figsize=(15,15)) #15 por 15 foi como comentamos na aula desse exemplo
saida[0].imshow(img)
saida[1].imshow(imgfinal)


print("Percebe-se que a técnica ensinada pelo mestrando consegue sim eliminar pequenos elementos, porém devido à propriedade da técnica, essa deixa pequenas manchas pretas que comprometem a visualização e compreensão da imagem.")

```
