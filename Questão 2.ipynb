```python
#Questão número 2: Aguçamento de Bordas
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Fig0359.tif')
edges = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

kernel = np.array([[0.0, -1.0, 0.0],
                   [-1.0, 4.0, -1.0],
                   [0.0, -1., 0.0],])

img2 =  cv2.filter2D(edges, -1,kernel)
img3 = cv2.hconcat([edges, img2])
#cv2.imshow('Imagens',img3)

cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#aguçamento

agu =  cv2.add(edges,img2)
imagem4 = cv2.hconcat([edges,agu])

#cv2.imshow('Imagens',imagem4)

#cv2.waitKey(0)
#cv2.destroyAllWindows()


plt.subplot(131), plt.imshow(img, "gray"), plt.title("Imagem Original")
plt.subplot(132), plt.imshow(agu, "gray"), plt.title("bordas realçada")
```
